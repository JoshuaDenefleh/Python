```python
#Bollinger Bnder

adyen['Open: 30 Day Mean'] = adyen['Open'].rolling(window=20).mean()
adyen['Upper'] = adyen['Open: 30 Day Mean'] + 2*adyen['Open'].rolling(window=20).std()
adyen['Lower'] = adyen['Open: 30 Day Mean'] - 2*adyen['Open'].rolling(window=20).std()
adyen[['Open','Open: 30 Day Mean','Upper','Lower']].plot(figsize=(16,6))
```

```python
returns_fig = sns.PairGrid(adyen.dropna())

returns_fig.map_upper(plt.scatter,color='purple')

returns_fig.map_lower(sns.kdeplot,cmap='cool_d')

returns_fig.map_diag(plt.hist,bins=30)
```
